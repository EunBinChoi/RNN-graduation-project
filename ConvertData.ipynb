```python
import numpy as np

xy = np.loadtxt("./data/2박수.csv", delimiter=',')
print(type(xy))
#xy=open('11.csv', 'r').readlines()
#print(xy.ndim)
print(xy.shape[0])

n = int(xy.shape[0] + xy.shape[0]/2)
print(n)
z = np.zeros(25)
k = 0
data = np.zeros( (n,25) )
print(data)
```

```python
for i in range(0, xy.shape[0],2):
    z[24] = int(xy[i+1, 24])
    print(z[24])
    
    for j in range(0, 25):
        data[k,j] = xy[i,j]
        data[k+1,j] = xy[i+1,j]
        data[k+2,j] = int(z[j])
        data[k+3,j] = int(z[j])
    k = k+4
print(data)    

with open('./data/out_2박수.csv', 'wb') as abc:
    np.savetxt(abc, data, delimiter=",")
```
