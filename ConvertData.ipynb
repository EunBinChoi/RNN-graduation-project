```python
import numpy as np

xy = np.loadtxt("222.csv", delimiter=',')
print(type(xy))
#xy=open('11.csv', 'r').readlines()
#print(xy.ndim)
print(xy.shape[0])

n = int(xy.shape[0] + xy.shape[0]/3)
z = np.zeros(25)
k = 0
data = np.zeros( (n,25) )
print(data)
```

```python
for i in range(0, xy.shape[0], 3):
    z[24] = int(xy[i+2, 24])
    print(z[24])
    for j in range(0, 25):
        data[k,j] = xy[i,j]
        data[k+1,j ] = xy[i+1,j]
        data[k+2,j ] = xy[i+2,j]
        data[k+3,j] = int(z[j])
    k = k+4
print(data)    

with open('output2.csv', 'wb') as abc:
    np.savetxt(abc, data, delimiter=",")
```
